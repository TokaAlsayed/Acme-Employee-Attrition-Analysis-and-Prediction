```python
#splitting
X = attr.drop('Attrition', axis = 1)
y = attr.Attrition
```

```python
#splitting data into train and test
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3)
print(X_train.shape) ; print(X_test.shape)
```

```python
#OrdinalEncoder 'BusinessTravel', 'Education', 'EnvironmentSatisfaction', 'JobInvolvement', 'JobSatisfaction','WorkLifeBalance',
#'PerformanceRating','RelationshipSatisfaction'
#binary gender  OverTime
#onehotencoder 'Department', 'EducationField', 'JobRole', 'MaritalStatus','BusinessTravel
```

```python
categorical_feature_ordinal= attr[[ 'Education', 'EnvironmentSatisfaction', 'JobInvolvement', 'JobSatisfaction','WorkLifeBalance',
'PerformanceRating','RelationshipSatisfaction','JobLevel','StockOptionLevel']]
categorical_feature_ordinal.head()
```

```python
categorical_feature=['Department', 'EducationField', 'JobRole', 'MaritalStatus','BusinessTravel']
```

```python
numerical_feature= ['Age','YearsWithCurrManager','YearsSinceLastPromotion','YearsInCurrentRole','YearsAtCompany','TrainingTimesLastYear','TotalWorkingYears',
                          'PercentSalaryHike','NumCompaniesWorked','MonthlyIncome','MonthlyRate','HourlyRate','DistanceFromHome','DailyRate']
```

```python
binary_features = ['Gender','OverTime']
```

```python
from sklearn.preprocessing import FunctionTransformer
def binary_transformer(X):
    return X.apply(lambda x: x.map({'Male': 0, 'Female': 1, 'Yes': 1, 'No': 0}))
binary_preprocessor = FunctionTransformer(binary_transformer)
```

```python
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))]) # One hot encoding
```

```python
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_feature),
        ('cat', categorical_transformer, categorical_feature),
        ('binary', binary_preprocessor, binary_features)], 
        remainder = 'passthrough')
```

```python
y_train.replace({'No':0,
                'Yes':1}, inplace = True)
y_test.replace({'No':0,
                'Yes':1}, inplace = True)
print(y_train.value_counts())
print(y_test.value_counts())
```

```python
models = {
    'random_forest': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', RandomForestClassifier())
    ]),
    'logistic_regression': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', LogisticRegression())
    ]),
    'decision_tree': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', DecisionTreeClassifier())
    ]),
    'gradient_boosting': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', GradientBoostingClassifier())
    ]),
    'xgboost': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'))
    ]),
    'adaboost': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', AdaBoostClassifier())
    ])
}

# Fit models
for name, pipeline in models.items():
    pipeline.fit(X_train, y_train)
   

```

```python
def predict(model):
    y_predict = model.predict(X_test)
    print('Accuracy: {:.2f}%'.format(accuracy_score(y_test, y_predict) * 100))
    print('Precision: {:.2f}%'.format(precision_score(y_test, y_predict, average='binary') * 100))
    print('Recall: {:.2f}%'.format(recall_score(y_test, y_predict, average='binary') * 100))
    print('F1_Score: {:.2f}%'.format(f1_score(y_test, y_predict, average='binary') * 100))

    # Generate the confusion matrix
    cm = confusion_matrix(y_test, y_predict)

    # Print the confusion matrix
    print("Confusion Matrix:")
    print("          Predicted:")
    print("             Neg    Pos")
    print("Actual Neg  {0: <6} {1: <6}".format(cm[0, 0], cm[0, 1]))
    print("       Pos  {0: <6} {1: <6}".format(cm[1, 0], cm[1, 1]))
```
